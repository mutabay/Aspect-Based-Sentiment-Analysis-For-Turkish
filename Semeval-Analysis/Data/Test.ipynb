```python
MODEL_NAME = 'dbmdz/bert-base-turkish-128k-cased'
MAX_LENGTH = 39  # We truncate anything after the 200-th word to speed up training

config = BertConfig.from_pretrained(MODEL_NAME)

tokenizer = BertTokenizer.from_pretrained(MODEL_NAME)

test_encodings = tokenizer(lem_text_2, truncation=True, padding='max_length',
                           max_length=MAX_LENGTH, return_tensors="tf")

test_dataset = tf.data.Dataset.from_tensor_slices(test_encodings['input_ids']).batch(1)
```

```python
re_designed_df = pd.read_csv('re-designed_train_df.csv')
re_designed_df.drop('Unnamed: 0', inplace=True, axis=1)
re_designed_df
```

```python
from sklearn.model_selection import train_test_split

label_names = re_designed_df.drop(["SID", "lemmatized"], axis=1).columns
labels = re_designed_df[label_names].values
texts = re_designed_df['lemmatized']

train_texts, test_texts, train_labels, test_labels = train_test_split(
    texts, labels, test_size=0.1, random_state=42
)
```

```python
def convert(label_string, label_onehot):
    labels = []
    for i, label in enumerate(label_string):
        if label_onehot[i]:
            labels.append(label)
    if len(labels) == 0:
        labels.append("neutral - RESTAURANT#GENERAL")
    return labels
```

```python
y=[]
result = []
print("144 sample prediction")
threshold = 0.4
i = 0
for prediction,a in zip(pred, test_dataset):
    print(i,'th loop')
    i += 1
    pred[pred>threshold]=1
    pred[pred<=threshold]=0
    convert_result = str(convert(label_names, prediction))
    print("prediction:", prediction, convert_result)
    y.append(prediction)
    result.append(convert_result)
```

```python
result
```

```python
import ast
# Removing strings
result = [ast.literal_eval(res) for res in result]
# Merging outer list with inner list
result = [j for i in result for j in i]
result
```

```python
import re

# Extracting polarities
regex_pol = r'[a-z]+\w'
polarities = [re.findall(regex_pol, res) for res in result]
polarities = [j for i in polarities for j in i]
len(polarities)
```

```python
regex_asp = r'[A-Z#]+\w'
aspects = [re.findall(regex_asp, res) for res in result]
aspects = [j for i in aspects for j in i]
aspects
```

```python
from tensorflow_addons.metrics import F1Score
from tensorflow_addons.optimizers import LazyAdam, Lookahead, MultiOptimizer
from tensorflow_addons.losses import SigmoidFocalCrossEntropy
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.metrics import AUC
from tensorflow.keras.optimizers import Adam

loss = BinaryCrossentropy(reduction = tf.keras.losses.Reduction.AUTO)

optimizers = [
    Lookahead(LazyAdam(1e-5), sync_period = 5),
    Adam(1e-5)
]
optimizers_and_layers = [(optimizers[0], loaded_model.layers[:2]),
                         (optimizers[1], loaded_model.layers[2:])]
optimizer = MultiOptimizer(optimizers_and_layers)

metrics = [
    F1Score(num_classes = train_labels.shape[1],
            average = 'micro',
            threshold = 0.4),

    "binary_accuracy",

    AUC(name="average_precision",
        curve="PR",
        multi_label=True)
]
loaded_model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

benchmarks = loaded_model.evaluate(
    test_dataset.batch(1),
    return_dict=True
)
```
